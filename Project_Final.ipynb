# Data Exploration!

## Question 1: What are the most common car makes and models in the dataset?

```python
common_car_makes_models = dd.groupby(['Make','Model']).size().reset_index(name='Count')
common_car_makes_models = common_car_makes_models.sort_values(by='Count', ascending=False)
top_5_common = common_car_makes_models.head(1)
print(top_5_common)
```

## Question 1.1: What are the most common car makes in the dataset?

```python
common_car_makes_models = dd.groupby(['Make']).size().reset_index(name='Count')
common_car_makes_models = common_car_makes_models.sort_values(by='Count', ascending=False)
top_5_common = common_car_makes_models.head(1)
print(top_5_common)
```

## Question 2: What is the distribution of car prices in the dataset?

```python
plt.hist(dd['MSRP'], bins=30, edgecolor='black')
plt.xlabel('Car Price')
plt.ylabel('Number of car')
plt.title('Distribution of Car Prices')
plt.show()
```

## Question 3: Which car brand has the highest average price?

```python
data['MSRP'] = data['MSRP'].replace({'\$': '', ',': ''}, regex=True).astype(float)
average_prices = data.groupby('Make')['MSRP'].mean().reset_index()
highest_average_price_brand = average_prices.loc[average_prices['MSRP'].idxmax()]
print(highest_average_price_brand['Make'],", Seems the Car brand with the highest average price:", )
```

## Question 4: What are the top 5 most expensive cars in the dataset?

```python
data_sorted = data.sort_values(by='MSRP', ascending=False)
top_5_expensive_cars = data_sorted.head(5)
print(top_5_expensive_cars[['Make', 'Model', 'MSRP']])
```

## Question 5: How does the mileage vary based on car make?

```python
mileage_by_make_model = data.groupby(['Make'])['MPG_City'].mean().reset_index()
sorted_mileage = mileage_by_make_model.sort_values(by='MPG_City', ascending=False)
plt.figure(figsize=(24, 6))
plt.bar(sorted_mileage['Make'], sorted_mileage['MPG_City'])
plt.xlabel('Car Make and Model')
plt.ylabel('Average Mileage (MPG_City)')
plt.title('Average Mileage by Car Make and Model')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Question 6: How does the fuel efficiency (MPG) vary across different car categories 

```python
mpg_by_category = data.groupby('Type')['MPG_City'].mean().reset_index()
sorted_mpg = mpg_by_category.sort_values(by='MPG_City', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(sorted_mpg['Type'], sorted_mpg['MPG_City'])
plt.xlabel('Car Category')
plt.ylabel('Average Fuel Efficiency (MPG)')
plt.title('Average Fuel Efficiency across Different Car Categories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Question 7: Are there any correlations between car features and car prices?

```python
selected_columns = ['Horsepower', 'EngineSize', 'MPG_City', 'MPG_Highway', 'MSRP']
data_selected = data[selected_columns]
correlations = data_selected.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Car Features and Price')
plt.show()
```

## Question 8: How does the car price vary based on the number of cylinders in the engine?

```python
average_price_by_year = data.groupby('Cylinders')['MSRP'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year['Cylinders'], average_price_by_year['MSRP'], marker='o')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Car Price')
plt.title('Change in Car Prices based on the number of cylinders ')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 9: How does the fuel efficiency (MPG) vary across different car Horsepower?

```python
average_price_by_year = data.groupby('MPG_City')['Horsepower'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year['Horsepower'], average_price_by_year['MPG_City'], marker='o')
plt.xlabel('Horsepower')
plt.ylabel('Milege')
plt.title('Change in Car Prices based on the number of cylinders ')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 10: How does the Type of the car vary across different Horsepower?

```python
sales_by_region = data.groupby('Type')['Horsepower'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%', colors=['green', 'blue', 'red', 'yellow', 'orange','grey'])
plt.axis('equal')
plt.title('Horsepower by Type')
plt.show()
```
